### Concatenate 3000 of the res

```python
df_concat_src = res.sample(n=4000)
res = res.drop(df_concat_src.index)
```

```python
res
```

```python
#create an empty df with column names
df_concat = pd.DataFrame(columns= res.columns)
```

```python
def concat_n_rows(n, df_concat_src, df_concat):
    randomrows = df_concat_src.sample(n=n)

    df_concat_src = df_concat_src.drop(randomrows.index)

    li = randomrows.values.tolist()

    new_txt = ''
    li_new = [0]*17
    for item in li:
        li_new = np.where(np.logical_or(li_new,item[2:]), 1, 0)
        new_txt = new_txt + ' '.join(item[1].split(' ')[:80])

    newrow = pd.DataFrame([[new_txt]+ li_new.tolist()],columns=['Text']+label_cols)

    df_concat = df_concat.append(newrow, ignore_index=False)
    return df_concat_src, df_concat
```

```python
for i in range(800):
    df_concat_src, df_concat = concat_n_rows(2, df_concat_src, df_concat)
    df_concat_src, df_concat = concat_n_rows(3, df_concat_src, df_concat)
```

```python
res = res.append(df_concat, ignore_index=True)
```

```python
res = res.append(df_all_zero_set,  ignore_index=True)
```

```python
from sklearn.model_selection import train_test_split

train, validate = train_test_split(res, test_size=0.2)
```
